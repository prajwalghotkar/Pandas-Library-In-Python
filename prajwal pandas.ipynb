```python
import pandas as pd
left=pd.DataFrame({'id':[1,2,3,4,5],'Name':['Alex','Amy','Allen','Alice','Ayoung'],'subject_id':['sub1','sub2','sub4','sub6','sub5']})
right=pd.DataFrame({'id':[1,2,3,4,5],'Name':['Billy','Brian','Bran','Bryce','Betty'],'subject_id':['sub2','sub4','sub3','sub6','sub5']})
print(left)
print('\n')
print(right)
```

```python
# inner join...
print(pd.merge(left,right,on=['id','subject_id']))
```

```python
print(pd.merge(left,right,on=['id','subject_id'],how='inner'))
```

```python
# left join....
print(pd.merge(left,right,on=['id','subject_id'],how="left"))
```

```python
# right join....
print(pd.merge(left,right,on=['id','subject_id'],how='right'))
```

```python
print(left)
print()
print(right)
```

```python
print(pd.merge(left,right,on=['id','subject_id'],how='outer'))
```

```python
print(pd.merge(left,right,on=['id','subject_id'],how="left"))
print()
print(pd.merge(left,right,on=['id','subject_id'],how='right'))
```

```python
print(pd.merge(left,right,on=['id','subject_id'],how='outer'))
```
